```python
def cost(x,y,m):
    return (1/len(x))*(((y-((x*m).sum(axis=1)))**2).sum())
```

```python
def getmc(x_train,y_train,learning_rate,m):
    scope_m=np.zeros(len(m))
   
    
    n=len(x_train)
    
    for i in range(n):
        
        for j in range(len(m)):
            scope_m[j]+=(-2/n)*((y_train[i]-(x_train[i]*m).sum())*x_train[i][j])
        
       
    new_m=m-learning_rate*scope_m
    
    return new_m
m=run()

#print(m)
test_data=np.loadtxt(r'C:\Users\user\Desktop\test.csv',delimiter=',')
df=pd.DataFrame(test_data)
    
features=df.columns
b=4
for a in features:
    df[b]=df[a]*df[a]
    b+=1
c=8
for a in features:
    df[c]=df[a]*df[a]*df[a]
    c+=1
    
    
test_data=np.array(df)
test_data=arg.transform(test_data)

amp=np.ones((test_data.shape[0],test_data.shape[1]+1))
amp[:,:-1]=test_data
test_data=amp

def pred(m,test_data):
    return (m*test_data).sum(axis=1)
        
    
a=pred(m,test_data)
np.savetxt('combinedcyclepowerplant.csv',a,delimiter=',')
print(a)
```
