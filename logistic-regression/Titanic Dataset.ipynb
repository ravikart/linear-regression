```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LogisticRegression

trainData = pd.read_csv('training_titanic_x_y_train.csv',delimiter=",")
testData = pd.read_csv('test_titanic_x_test.csv',delimiter=",")

#trainData = pd.concat([trainData,testData],axis=0)


trainData = trainData.drop(['Ticket','Cabin'],axis=1)
trainData['Title'] = trainData.Name.str.extract(' ([A-Za-z]+)\.',expand=False)
testData = testData.drop(['Ticket','Cabin'],axis=1)


trainData['Title'] = trainData['Title'].replace(['Lady','Capt','Col','Countess','Don','Dr','Major','Rev','Sir','Jonkheer','Dona'],'Rare')
trainData['Title'] = trainData['Title'].replace('Mlle','Miss')
trainData['Title'] = trainData['Title'].replace('Ms','Miss')
trainData['Title'] = trainData['Title'].replace('Mme','Mrs')

title_mapping = {"Mr": 1, "Miss": 2, "Mrs": 3, "Master": 4, "Rare": 5}
sex_mapping = {'female':1,'male':0}

trainData['Title'] = trainData['Title'].map(title_mapping)
trainData['Title'] = trainData['Title'].fillna(0)

trainData = trainData.drop(['Name'],axis=1)

trainData['Sex'] = trainData['Sex'].map(sex_mapping).astype(int)

trainData.groupby('Pclass').mean()['Age'].round()
            
mean_class1 = trainData.groupby('Pclass').mean()['Age'].round().loc[1]
mean_class2 = trainData.groupby('Pclass').mean()['Age'].round().loc[2]
mean_class3 = trainData.groupby('Pclass').mean()['Age'].round().loc[3]

trainData.loc[trainData['Pclass']==1,'Age'] = trainData.loc[trainData['Pclass']==1,'Age'].fillna(value=mean_class1)
trainData.loc[trainData['Pclass']==2,'Age'] = trainData.loc[trainData['Pclass']==2,'Age'].fillna(value=mean_class2)
trainData.loc[trainData['Pclass']==3,'Age'] = trainData.loc[trainData['Pclass']==3,'Age'].fillna(value=mean_class3)


print(trainData.info())

sex = pd.get_dummies(trainData['Sex'],drop_first=True)
embark = pd.get_dummies(trainData['Embarked'],drop_first=True)


trainData = pd.concat([trainData,sex,embark],axis=1)

trainData.drop(['Sex','Embarked'],axis=1,inplace=True)

print(trainData.info())

```
