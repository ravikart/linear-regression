```python
from sklearn.linear_model import LogisticRegression
import numpy as np
import pandas as pd
```

```python
titanic_train=pd.read_csv("Downloads/titanic_train.csv",delimiter=',')
titanic_test=pd.read_csv("Downloads/0000000000002429_test_titanic_x_test.csv")
```

```python
titanic_train.head()
```

```python
titanic_test.head()
```

```python
titanic_train.drop(["Name","Ticket","Cabin"],axis=1,inplace=True)
```

```python
titanic_train
```

```python
titanic_test.drop(["Name","Ticket","Cabin"],axis=1,inplace=True)
```

```python
titanic_test
```

```python
titanic_train["Age"].fillna(titanic_train["Age"].mean(),inplace=True)
```

```python
titanic_test["Age"].fillna(titanic_train["Age"].mean(),inplace=True)
```

```python
def Sex_num(sex):
    if(sex=='male'):
        return 0
    elif(sex=="female"):
        return 1
```

```python
titanic_train["Sex"]=titanic_train["Sex"].apply(Sex_num)
```

```python
titanic_train
```

```python
titanic_test["Sex"]=titanic_test["Sex"].apply(Sex_num)
titanic_test
```

```python
def embarked(emp):
    if(emp=='S'):
        return 0
    elif(emp=='Q'):
        return 1
    elif(emp=='C'):
        return 2
    else:
        return 0
```

```python
titanic_test["Embarked"]=titanic_test["Embarked"].apply(embarked)
titanic_test
```

```python
titanic_train["Embarked"]=titanic_train["Embarked"].apply(embarked)
titanic_train
```

```python
x_train=titanic_train.drop("Survived",axis=1)
```

```python
y_train=titanic_train["Survived"]
```

```python
alg=LogisticRegression()
```

```python
alg.fit(x_train,y_train)
```

```python
for i in alg.predict(x_train)-y_train:
    print(i)
```

```python
alg.score(x_train,y_train)
```

```python
y_test=alg.predict(titanic_test)
```

```python
print(y_test)
```
